```
Given an array where elements are sorted in ascending order, convert it to a height balanced BST.

For this problem, a height-balanced binary tree is defined as a binary tree in which the depth of the two subtrees of every node never differ by more than 1.

Example:

Given the sorted array: [-10,-3,0,5,9],

One possible answer is: [0,-3,9,-10,null,5], which represents the following height balanced BST:

      0
     / \
   -3   9
   /   /
 -10  5


```

````python
class TreeNode(object):
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
    def __str__(self):
        def printT(n, prefix, cprefix):
            if not n:
                return ''
            s = ''
            s += prefix
            s += str(n.val)
            s += "\n"
            s += printT(n.left, cprefix + "|- ", cprefix + "|  ")
            s += printT(n.right, cprefix + "|- ", cprefix + "|  ")
            return s

        a = printT(self, "", "")
        return a
    def __eq__(self, other):
        return str(self) == str(other)


def makeTree(vals):
    if not vals:
        return None
    r = TreeNode(-1)
    q = [r]
    #import pdb; pdb.set_trace()

    while vals:
        v = vals.pop(0)
        n = q.pop(0)
        if not v:
            continue
        n.val = v
        
        # TBD: valsにNoneで位置が含まれているので、ノードも位置が必要
        n.left = TreeNode(None)
        n.right = TreeNode(None)
        
        q.append(n.left)
        q.append(n.right)
    
    # TreeNode(None)を消す
    def dfs(n):
        if n.left and not n.left.val:
            n.left = None
        else:
            dfs(n.left)
            
        if n.right and not n.right.val:
            n.right = None
        else:
            dfs(n.right)
        return n
    
    r = dfs(r)
    return r        
a=makeTree([2,1,3])
#print(a)
a=makeTree([5,1,4,None,None,3,6])
#print(a)
````

````python
# ok
class Solution(object):
    def sortedArrayToBST(self, nums):
        def dfs(nums):
            if not nums:
                return None
            idx = len(nums)/2
            
            root = TreeNode(nums[idx])
            root.left = dfs(nums[0:idx])
            root.right = dfs(nums[idx+1:])
            
            return root
        return dfs(nums)
            
        

    
````
