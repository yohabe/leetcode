```
104. Maximum Depth of Binary Tree
Easy

Given a binary tree, find its maximum depth.

The maximum depth is the number of nodes along the longest path from the root node down to the farthest leaf node.

Note: A leaf is a node with no children.

Example:

Given binary tree [3,9,20,null,null,15,7],

    3
   / \
  9  20
    /  \
   15   7

return its depth = 3.
```

````python
class TreeNode(object):
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
    def __str__(self):
        def printT(n, prefix, cprefix):
            if not n:
                return ''
            s = ''
            s += prefix
            s += str(n.val)
            s += "\n"
            s += printT(n.left, cprefix + "|- ", cprefix + "|  ")
            s += printT(n.right, cprefix + "|- ", cprefix + "|  ")
            return s

        a = printT(self, "", "")
        return a

def makeTree(vals):
    if not vals:
        return None
    r = TreeNode(-1)
    q = [r]
    #import pdb; pdb.set_trace()

    while vals:
        v = vals.pop(0)
        n = q.pop(0)
        if not v:
            continue
        n.val = v
        
        # TBD: valsにNoneで位置が含まれているので、ノードも位置が必要
        n.left = TreeNode(None)
        n.right = TreeNode(None)
        
        q.append(n.left)
        q.append(n.right)
    
    # TreeNode(None)を消す
    def dfs(n):
        if n.left and not n.left.val:
            n.left = None
        else:
            dfs(n.left)
            
        if n.right and not n.right.val:
            n.right = None
        else:
            dfs(n.right)
        return n
    
    r = dfs(r)
    return r        
a=makeTree([2,1,3])
#print(a)
a=makeTree([5,1,4,None,None,3,6])
#print(a)
````

````python
# recursive
class Solution(object):
    def maxDepth(self, root):
        def dfs(n):
            if not n:
                return 0
            return 1 + max(dfs(n.left), dfs(n.right))
        if not root:
            return 0
        return dfs(root)
````

````python
# iterate
class Solution(object):
    def maxDepth(self, root):
        def dfs(n):
            stack = [(n, 1)]
            ans = 0
            while stack:
                n, depth = stack.pop()
                ans = max(ans, depth)
                if n.left:
                    stack.append((n.left, depth+1))
                if n.right:
                    stack.append((n.right, depth+1))
            return ans
        if not root:
            return 0
        return dfs(root)
                                
            

````
