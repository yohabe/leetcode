```python
class ListNode(object):
    def __init__(self, x):
        self.val = x
        self.next = None
    
    def __str__(self):
        s = []
        c = self
        while c:
            s.append(c.val)
            c = c.next
        return '->'.join(s)
    
    def __eq__(self, other):
        return str(self) == str(other)
        
        
# str: 1->3->3
def makeList(arg):
    if type(arg) is str:
        a = str.split('->')
    else:
        a = arg
    h = cur = ListNode(0)
    for i in a:
        t = ListNode(i)
        cur.next = t
        cur = t
    return h.next
```

```python
class TreeNode(object):
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
    def __str__(self):
        def printT(n, prefix, cprefix):
            if not n:
                return ''
            s = ''
            s += prefix
            s += str(n.val)
            s += "\n"
            s += printT(n.left, cprefix + "|- ", cprefix + "|  ")
            s += printT(n.right, cprefix + "|- ", cprefix + "|  ")
            return s

        a = printT(self, "", "")
        return a
    def __eq__(self, other):
        return str(self) == str(other)


def makeTree(vals):
    if not vals:
        return None
    r = TreeNode(-1)
    q = [r]
    #import pdb; pdb.set_trace()

    while vals:
        v = vals.pop(0)
        n = q.pop(0)
        if not v:
            continue
        n.val = v
        
        # TBD: valsにNoneで位置が含まれているので、ノードも位置が必要
        n.left = TreeNode(None)
        n.right = TreeNode(None)
        
        q.append(n.left)
        q.append(n.right)
    
    # TreeNode(None)を消す
    def dfs(n):
        if n.left and not n.left.val:
            n.left = None
        else:
            dfs(n.left)
            
        if n.right and not n.right.val:
            n.right = None
        else:
            dfs(n.right)
        return n
    
    r = dfs(r)
    return r        
a=makeTree([2,1,3])
#print(a)
a=makeTree([5,1,4,None,None,3,6])
#print(a)
```
